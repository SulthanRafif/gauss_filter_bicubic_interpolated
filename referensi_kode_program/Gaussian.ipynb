# Gaussian Filter

https://www.geeksforgeeks.org/apply-a-gauss-filter-to-an-image-with-python/

Gaussian Filter merupakan low pass filter yang digunakan untuk mengurangi noise (high frequency components) dan bagian blur pada  citra. Filter ini mengimplementasikan Odd Sized Symmetric Kernel atau Kernel Simetris Berukuran Ganjil yang diterapkan pada setiap pixel yang ditentukan untuk mendapatkan efek yang diinginkan. Kernel ini tidak diperuntukkan untuk melakukan perubahan warna pada citra secara drastis (contoh untuk menampilkan tepi/edges) dikarenakan (pre-memory). Gaussian Filter merupakan penerapan dari pendekatan Fungsi Gaussian.

https://www.youtube.com/watch?v=afpIGjIpjT8

Digunakan untuk proses penghalusan (smoothing), pengaburan (blurring), menghilangkan detail dan menghilangkan noise. Nilai intensitas setiap pixel diganti dengan rata-rata nilai pembobotan dari setiap pixel-pixel tetangganya dan pixel itu sendiri. Pixel tetangga adalah pixel yang berada di sekeliling pixel yang dimaksud. Jumlah tetangga yang dilibatkan tergantung pada filter yang dirancang. Bergantung pada kernelnya sebagai filter.

Langkah-langkah:
- Menentukan matriks kernel (3x3, 5x5, 7x7 berukuran ganjil)
Semakin kecil nilai kernel, maka akan semakin banyak nilai pixel pusat kernel yang berubah nilainya sehingga perubahan citra semakin tampak. Dan sebaliknya. 

- Menghitung nilai elemen matriks kernel
Menggunakan rumus
${g(x.y) = e^{-{x^2 + y^2 \over 2 σ^2}}}$

σ (standar deviasi) = 2

e (konstanta euler (2.718281828))

- Normalisasi matriks filter gaussian

Membulatkan masing-masing nilai matriks filter yang telah didapat, karena nilai intensitas pixel pada citra digital bernilai bulat. Dengan rumus:

${c = {|{1 \over min(g(x.y))}|}}$

Dicari inilai terkecil agar jumlah nilai-nilai bobot sama dengan satu. Selanjutnya C akan dikalikan dengan nilai setiap elemen pada kernel.

- Mengalikan matriks kernel dengan nilai normalisasi

${g(x.y) = c . {e^{-{x^2 + y^2 \over 2 σ^2}}}}$

Setelah dilakukan perkalian matriks kernel dengan nilai normalisasi dilakukan penjumlahan semua elemen nilai pembobot g(x,y) sebagai pembagi matriks filter. Dengan tujuan nilai intensitas citra tetap seperti semula dan menjaga agar nilai baru tidak berada di luar batas nilai grayscale.

- Proses konvolusi
Konvolusi pada citra sering disebut sebagai konvolusi 2 Dimensi. Konvolusi 2 Dimensi didefinisikan sebagai proses untuk memperoleh suatu pixel berdasarkan nilai pixel itu sendiri dan tetangganya dengan melibatkan matriks kernel.

${h(x,y) = f(x,y) \times g(x,y)}$

## Proses untuk menampilkan filter gauss
- Tahapan pertama terlebih dahulu mendefinisikan ukuran dari kernel/matrix yang digunakan untuk mengurangi (demising) citra. 
- Ukuran umum yang diterapkan secara general berupa angka ganjil, yaitu merupakan hasil keseluruhan yang dihitung dari pixel pusat.
- Kemudian kernel merupakan matrix simetris dan memiliki jumlah baris dan kolom yang sama. 
- Nilai didalam kernel dihitung menggunakan Fungsi Gaussian, dengan rumus:

${g(x.y) = {1 \over 2 \pi σ^2}{e^{-{x^2 + y^2 \over 2 σ^2}}}}$

Implementasi Gaussian Kernel di Python

```python
# ImageFilter for using filter() function
from PIL import Image, ImageFilter

# Opening the image
# (R prefixed to string in order to deal with '\' in paths)
image = Image.open('./BrainDataset/testA/ct10.png')

# Blurrinng image by sending the ImageFilter.
# GaussianBlur predefined kernel argument, menggunakan ukuran kernel 5x5 untuk efek blur
image = image.filter(ImageFilter.GaussianBlur)

# Displaying the Image
image.save('./HasilGaussian/output.png')
```

```python
from PIL import Image, ImageFilter
  
image = Image.open('./BrainDataset/testA/ct10.png')
  
# Cropping the image 
smol_image = image.crop((0, 300, 0, 300))
  
# Blurring on the cropped image
blurred_image = smol_image.filter(ImageFilter.GaussianBlur)
  
# Pasting the blurred image on the original image
image.paste(blurred_image, (0,0))
  
# Displaying the image
image.save('./HasilGaussian/output_partial_3.png')
```
