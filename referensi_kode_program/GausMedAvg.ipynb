```python
def gausblur(image,blur,m):
    image = cv2.GaussianBlur(image,(m,m),blur)
    return image
def medblur(image,shift):
    image = cv2.medianBlur(image,shift)
    return image
def avgfilter(image,shift):
    image = cv2.blur(image,(shift,shift))
    return image
```
